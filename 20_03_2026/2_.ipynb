```python
from openai import OpenAI

client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "Explain what LangChain is in simple terms"}
    ],
    temperature=0.7
)

print(response.choices[0].message.content)
```

____

- temperature — What it Actually Controls
  - temperature controls the randomness of the model’s token selection.
    -  0.0 → deterministic (very predictable)
    -  0.2–0.5 → controlled, focused
    -  0.7 → balanced (default-ish)
    -  1.0+ → creative, diverse, sometimes messy



- temperature modifies that probability distribution:

  - Low temperature (0.1)
    - Sharp distribution → top tokens dominate
    - Output is repetitive but reliable
  - High temperature (1.0)
    - Flatter distribution → more tokens considered
    - Output is diverse but less stable

____


#### Prompt

```python
from openai import OpenAI

client = OpenAI()

def run_prompt(messages, temperature=0.7):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=temperature
    )
    return response.choices[0].message.content

                         - - -                                         - - -                                              - - -

#### Basic (zero-shot)

```python
messages = [
    {"role": "user", "content": "Explain transformers in simple terms"}
]

print(run_prompt(messages))

-  -  -                     -  -  -                       -  -  -                                                          -  - - 

#### Structured Prompt

```python
messages = [
    {
        "role": "user",
        "content": """
Explain transformers in simple terms.

Format:
1. Definition
2. Key idea
3. Example
"""
    }
]

print(run_prompt(messages))
```

___

#### Role-Based Prompt (System + User)

```python
messages = [
    {"role": "system", "content": "You are a helpful AI tutor who explains concepts simply."},
    {"role": "user", "content": "Explain transformers"}
]

print(run_prompt(messages))

___

#### Few-Shot Prompt
```python
messages = [
    {"role": "user", "content": "Explain neural networks"},
    {"role": "assistant", "content": "Neural networks are systems inspired by the brain that learn patterns from data."},

    {"role": "user", "content": "Explain decision trees"},
    {"role": "assistant", "content": "Decision trees are models that split data into branches to make decisions."},

    {"role": "user", "content": "Explain transformers"}
]

print(run_prompt(messages))

___

1. Core Types (what you just learned)

These are the primitives:

- Zero-shot
- Structured
- Role-based
- Few-shot

Everything else builds on top of these.

## Reasoning-Oriented Prompts (Very Important)

#### 1.Chain-of-Thought (CoT)
- Force the model to reason step-by-step.

```python
messages = [
    {"role": "user", "content": "Solve: If a train travels 60 km in 1 hour, how far in 3 hours? Think step by step."}
]
```

#### 2.Self-Consistency Prompting

- Ask multiple reasoning paths → pick best (done programmatically)
- Generate 5 answers → choose most common


#### 3.ReAct (Reason + Act)

- Used heavily in agents (LangChain later)
```python
Thought: I need to search
Action: search("LangChain")
Observation: ...
```
