- LangChain memory is a state management layer that:
   - Stores past interactions (chat history, variables, summaries)
   - Decides what to keep / drop
   - Injects relevant parts into future prompts automatically

####  Types of Memory in LangChain

- Buffer Memory(ConversationBufferMemory)
   - Injects full history (can hit token limits)
-  Window Memory (ConversationBufferWindowMemory)
   - Keep last N messages → Inject
- Summary Memory(ConversationSummaryMemory)
   - Uses an LLM to compress history:
   - Efficient for long conversations
- Vector Memory (semantic memory)
   - Stores interactions in embeddings
   - Retrieves only relevant past info, not everything
   - This is where memory starts overlapping with RAG

-  Entity Memory
   - Extract entities → store as key-value
