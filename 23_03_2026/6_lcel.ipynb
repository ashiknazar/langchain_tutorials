### LCEL (LangChain Expression Language)

- It’s a new way to build chains using simple, composable pipelines.
- Instead of writing complex classes, you just connect components like Lego blocks.

### Core Idea

- LCEL uses this operator:
   - `|`
   - Read it as: “then pass output to next step”

#### Basic Example

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain.prompts import ChatPromptTemplate

llm = ChatOpenAI(model="gpt-4o-mini")

prompt = ChatPromptTemplate.from_template("Explain {topic} in 2 lines")

chain = prompt | llm | StrOutputParser()

print(chain.invoke({"topic": "Kafka"}))

___

Input → Prompt → LLM → Parser → Output

___
___

### Why LCEL Was Introduced

- Old LangChain (Chains API):
```python
LLMChain(llm=llm, prompt=prompt)

- Problems:
  - Rigid
  - Hard to customize
  - Hard to debug

- LCEL:
  - prompt | llm | parser

___
___

#### LCEL Building Blocks(LangchainExpressionLanguage)

- Everything in LCEL is a Runnable

| Component | Example                |
| --------- | ---------------------- |
| Prompt    | `PromptTemplate`       |
| Model     | `ChatOpenAI`           |
| Parser    | `StrOutputParser`      |
| Function  | Custom Python function |


___

####  You Can Add Custom Logic Anywhere
```python
def to_upper(text):
    return text.upper()

chain = prompt | llm | StrOutputParser() | to_upper

___

#### Parallel Execution

```python
from langchain_core.runnables import RunnableParallel

chain = RunnableParallel({
    "definition": prompt | llm,
    "summary": prompt | llm
})

___

#### Branching (Conditional Logic)

```python
from langchain_core.runnables import RunnableBranch

def is_question(x):
    return "?" in x

chain = RunnableBranch(
    (is_question, llm),
    StrOutputParser()
)
