- With prompt
```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in exactly 2 lines for a beginner."
)

formatted_prompt = prompt.format(topic="Apache Kafka")

llm.invoke(formatted_prompt)
```

### Types of Prompts in LangChain

#### 1.PromptTemplate (Basic)

- Used for simple text formatting.
```python
PromptTemplate(
    input_variables=["name"],
    template="Hello {name}, explain Spark in 3 bullet points."
)
```

#### 2.ChatPromptTemplate

- Used with chat models like ChatOpenAI.
```python
from langchain.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a strict data engineer."),
    ("user", "Explain Kafka")
])

llm.invoke(prompt.format_messages())
```

- This introduces roles:
  - system → controls behavior
  - user → actual question
  - assistant → optional examples

- System message = personality of LLM
  - “You are a strict teacher” → harsh answers
  - “You are a friendly tutor” → soft explanations
  - “You are a JSON generator” → structured output
