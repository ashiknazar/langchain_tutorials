 - CSVs

```bash
 Document(page_content="...", metadata={"source": "..."})
```

#### 3. Load Different Types of Documents
- Load a PDF


```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("your_file.pdf")
docs = loader.load()

print(len(docs))
print(docs[0].page_content[:500])
```

___

- B. Load a Text File

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("my_notes.txt")
docs = loader.load()

```

___

- C. Load a Web Page

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://scikit-learn.org/stable/")
docs = loader.load()
```

___

- D. Load a Folder of Files
```python
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader("data/", glob="**/*.txt")
docs = loader.load()
```

#### ✂️ 4. Chunking (Splitting Documents into Smaller Pieces)

- Because LLMs cannot process large PDFs at once.
- The standard tool is `RecursiveCharacterTextSplitter`.

#### Basic chunking
```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)

print(len(chunks))
print(chunks[0].page_content)
```

#### chunking tips

- Chunk size
  - 300–500 tokens → for general Q&A
  - 800–1500 tokens → for technical documents
  - 2000 tokens → for LLaMA 70B or GPT-4-level models
- Overlap
  - Overlap helps keep context between chunks.
    - 50 for small chunks
    - 100–200 for larger ones

### Example: Full Code to Load & Chunk a PDF

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load PDF
loader = PyPDFLoader("my_file.pdf")
docs = loader.load()

# Chunk
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)

print("Pages:", len(docs))
print("Chunks:", len(chunks))
print("Sample chunk:\n", chunks[0].page_content)
```
