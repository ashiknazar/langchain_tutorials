### What is LangChain? (Beginner Friendly Explanation)
 
- LangChain is a framework for building LLM-powered applications such as:
   - Chatbots
   - Question answering systems
   - RAG (Retrieval-Augmented Generation) apps
   - Agents & tools (LLMs calling functions)
   - Pipelines using multiple models and steps

___

| Problem            | Without LangChain                 | With LangChain                 |
| ------------------ | --------------------------------- | ------------------------------ |
| Load documents     | You write custom code             | `DocumentLoaders`              |
| Chunk & clean text | Write preprocessing manually      | `TextSplitters`                |
| Embed & store      | Write embedding + vector DB logic | `Embeddings + VectorStores`    |
| RAG pipeline       | Manually connect steps            | `RetrievalChain`               |
| Agents & tools     | Hard to structure                 | Few lines with `AgentExecutor` |


___

### LangChain Simple Architecture

Document → Split → Embed → Store → Retrieve → LLM → Answer


- This is the core of RAG.

### Step 2: Create a Chat Prompt Template

```python
from langchain.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a friendly scikit-learn tutor."),
    ("human", "{question}")
])

response = llm.invoke(prompt.format(question="What is cross-validation?"))
print(response)
```

### Step 3: Create a Chain (LLM + Prompt)
```python
from langchain.chains import LLMChain

chain = LLMChain(
    llm=llm,
    prompt=prompt
)

print(chain.invoke({"question": "Explain PCA in simple terms"}))
```
